```python
df_sales=pd.read_csv(r"C:\Users\Rajan\Downloads\sales_data_final.csv")
df_sales.head()
```

```python
#create a new column and store some derived data into that
#create profit_flag with the data of +ve and -ve where profit is +ve or -ve
# below is a user defined funtion // we can also use predefined function like 'len' for customer lenght

def profit_flag(a):
    if a>0:
        return 'positive'
    else:
        return 'negative'
profit_flag(-45)
```

```python
profit_flag(88)
```

```python
#create a new parallel column with thie function profit_flag (like map, filter and reduce which were aplied on List)
df_sales['flag_profit']=df_sales['profit'].apply(profit_flag)
df_sales
```

```python
#try to create one more column to store the lenght of the customer name
df_sales['length_customer']= df_sales['customer_name'].apply(len)
df_sales.head()
```

```python
#create a new column to derive quantity is low or medium or high
def quantity_flag(a):
    if a<10:
        return 'low'
    elif a >10 and a <20:
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['flag_quqantity'] = df_sales['quantity'].apply(quantity_flag)
df_sales.head()
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
df_sales.head()
```
