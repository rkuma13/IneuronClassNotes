```python
x = np.linspace(-1,6,30)
x
```

```python
y = np.linspace(-1,6,30)
y
```

```python
def sin_fun(x,y):
    return np.sin(np.sqrt(x**2 , y**2))
```

```python
z = sin_fun(x,y)
z
```

```python
z1=x+y
```

```python
ax = pit.axes(projection = '3d')
ax.plot3D(x,y,z)
```

```python
ax = pit.axes(projection = '3d')
ax.plot3D(x,y,z1)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
ax = pit.axes(projection = '3d')
ax.plot3D(df3['sepal_length'],df3['sepal_width'],df3['petal_length'])
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length')
```
